```python
"""
=============================================================
POTATO BLIGHT DISEASE - GOOGLE COLAB VERSION
Chạy từng cell theo thứ tự
=============================================================
"""

# ============================================================
# CELL 1: CÀI ĐẶT & IMPORT
# ============================================================
!pip install -q kaggle

import os, shutil, random
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score
from sklearn.utils.class_weight import compute_class_weight
from collections import Counter

print("✅ TensorFlow:", tf.__version__)
print("✅ GPU:", tf.config.list_physical_devices('GPU'))

# ============================================================
# CELL 2: TẢI DATASET TỪ KAGGLE
# ============================================================
# Bước 1: Upload file kaggle.json
# Vào https://www.kaggle.com → Account → API → "Create New API Token"
# → tải về kaggle.json → upload lên Colab ở bước dưới

from google.colab import files
print("📁 /content/potato_blight_COLAB.py")
uploaded = files.upload()

# Bước 2: Cấu hình kaggle
os.makedirs('/root/.kaggle', exist_ok=True)
os.system('cp kaggle.json /root/.kaggle/kaggle.json')
os.system('chmod 600 /root/.kaggle/kaggle.json')

# Bước 3: Tải dataset PlantVillage (chứa ảnh khoai tây)
print("\n📥 Đang tải dataset...")
os.system('kaggle datasets download -d arjuntejaswi/plant-village -p /content/')
os.system('unzip -q /content/plant-village.zip -d /content/PlantVillage_raw')
print("✅ Tải xong!")

# Kiểm tra cấu trúc
print("\n📂 Cấu trúc thư mục gốc:")
for item in sorted(os.listdir('/content/PlantVillage_raw/PlantVillage')):
    if 'Potato' in item:
        path = f'/content/PlantVillage_raw/PlantVillage/{item}'
        count = len(os.listdir(path))
        print(f"   {item}: {count} ảnh")

# ============================================================
# CELL 3: CHUẨN BỊ CẤU TRÚC THƯ MỤC TRAIN/TEST
# ============================================================
RAW_DIR     = '/content/PlantVillage_raw/PlantVillage'
DATASET_DIR = '/content/PlantVillage'

CLASS_MAP = {
    'Potato___Early_blight': 'Early_Blight',
    'Potato___Late_blight':  'Late_Blight',
    'Potato___healthy':      'Healthy',
}

random.seed(42)

# Xóa nếu đã tồn tại
if os.path.exists(DATASET_DIR):
    shutil.rmtree(DATASET_DIR)

print("🔄 Đang tổ chức dữ liệu train/test...")
for src_name, dst_name in CLASS_MAP.items():
    src_path = os.path.join(RAW_DIR, src_name)
    if not os.path.exists(src_path):
        print(f"⚠️  Không tìm thấy: {src_path}")
        continue

    imgs = [f for f in os.listdir(src_path) if f.lower().endswith(('.jpg','.jpeg','.png'))]
    random.shuffle(imgs)
    split = int(len(imgs) * 0.85)   # 85% train, 15% test

    for subset, img_list in [('train', imgs[:split]), ('test', imgs[split:])]:
        dst = os.path.join(DATASET_DIR, subset, dst_name)
        os.makedirs(dst, exist_ok=True)
        for img in img_list:
            shutil.copy(os.path.join(src_path, img), os.path.join(dst, img))

    print(f"   ✅ {dst_name}: {split} train / {len(imgs)-split} test")

print("\n📁 Cấu trúc dataset hoàn tất:")
for split in ['train', 'test']:
    for cls in ['Early_Blight', 'Late_Blight', 'Healthy']:
        p = os.path.join(DATASET_DIR, split, cls)
        if os.path.exists(p):
            print(f"   {split}/{cls}: {len(os.listdir(p))} ảnh")

# ============================================================
# CÂU 1: TIỀN XỬ LÝ DỮ LIỆU
# ============================================================
# CELL 4
print("\n" + "="*60)
print("CÂU 1: TIỀN XỬ LÝ DỮ LIỆU")
print("="*60)

IMG_SIZE    = (224, 224)
BATCH_SIZE  = 32
EPOCHS      = 50
NUM_CLASSES = 3
SEED        = 42
CLASS_NAMES = ['Early_Blight', 'Late_Blight', 'Healthy']

# Generator có augmentation cho train + validation
train_val_datagen = ImageDataGenerator(
    rescale=1.0/255.0,           # Chuẩn hóa pixel [0,1]
    rotation_range=20,
    width_shift_range=0.15,
    height_shift_range=0.15,
    shear_range=0.1,
    zoom_range=0.15,
    horizontal_flip=True,
    fill_mode='nearest',
    validation_split=0.2         # 20% làm validation
)

# Generator cho test (chỉ chuẩn hóa, không augment)
test_datagen = ImageDataGenerator(rescale=1.0/255.0)

# Train (80% của train folder)
train_generator = train_val_datagen.flow_from_directory(
    os.path.join(DATASET_DIR, 'train'),
    target_size=IMG_SIZE,        # Chuẩn hóa kích thước 224x224
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='training',
    seed=SEED
)

# Validation (20% của train folder)
val_generator = train_val_datagen.flow_from_directory(
    os.path.join(DATASET_DIR, 'train'),
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    subset='validation',
    seed=SEED
)

# Test
test_generator = test_datagen.flow_from_directory(
    os.path.join(DATASET_DIR, 'test'),
    target_size=IMG_SIZE,
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

print(f"\n✅ Tỷ lệ chia dữ liệu:")
print(f"   Train      : {train_generator.samples} ảnh  (~68% tổng)")
print(f"   Validation : {val_generator.samples} ảnh  (~17% tổng)")
print(f"   Test       : {test_generator.samples} ảnh  (~15% tổng)")
print(f"   Classes    : {train_generator.class_indices}")
print(f"   Pixel range: [0, 1]")

# Xử lý mất cân bằng
labels       = train_generator.classes
label_counts = Counter(labels)
class_weights_arr = compute_class_weight('balanced', classes=np.unique(labels), y=labels)
class_weight_dict = dict(enumerate(class_weights_arr))

print(f"\n⚖️  Class Weights (xử lý mất cân bằng):")
for i, name in enumerate(train_generator.class_indices.keys()):
    print(f"   {name}: {class_weight_dict[i]:.4f}")

# Biểu đồ phân phối lớp
fig, ax = plt.subplots(figsize=(8, 4))
cls_list   = list(train_generator.class_indices.keys())
cls_counts = [label_counts[i] for i in range(NUM_CLASSES)]
colors = ['#e67e22', '#c0392b', '#27ae60']
bars = ax.bar(cls_list, cls_counts, color=colors, edgecolor='white', linewidth=1.5)
for bar, val in zip(bars, cls_counts):
    ax.text(bar.get_x()+bar.get_width()/2, bar.get_height()+10,
            str(val), ha='center', fontsize=12, fontweight='bold')
ax.set_title('Phân phối số lượng ảnh theo lớp (Train)', fontsize=13, fontweight='bold')
ax.set_ylabel('Số lượng ảnh')
ax.set_xlabel('Lớp bệnh')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.savefig('/content/class_distribution.png', dpi=150)
plt.show()

# Hiển thị một số ảnh mẫu
fig, axes = plt.subplots(2, 6, figsize=(16, 6))
batch_imgs, batch_lbls = next(train_generator)
cls_names_list = list(train_generator.class_indices.keys())
for i in range(12):
    ax = axes[i//6][i%6]
    ax.imshow(batch_imgs[i])
    ax.set_title(cls_names_list[np.argmax(batch_lbls[i])], fontsize=9)
    ax.axis('off')
plt.suptitle('Ảnh mẫu sau tiền xử lý (pixel [0,1])', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/sample_images.png', dpi=150)
plt.show()
print("✅ Tiền xử lý hoàn tất!")

# ============================================================
# CÂU 2: THIẾT KẾ KIẾN TRÚC CNN
# ============================================================
# CELL 5
print("\n" + "="*60)
print("CÂU 2: THIẾT KẾ KIẾN TRÚC CNN")
print("="*60)

def build_potato_cnn(input_shape=(224, 224, 3), num_classes=3):
    inputs = keras.Input(shape=input_shape, name='input_image')

    # BLOCK 1: Conv×2 → BN → ReLU → MaxPool → Dropout
    x = layers.Conv2D(32, (3,3), padding='same')(inputs)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(32, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)          # 112×112
    x = layers.Dropout(0.25)(x)

    # BLOCK 2: Conv×2 → BN → ReLU → MaxPool → Dropout
    x = layers.Conv2D(64, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(64, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)          # 56×56
    x = layers.Dropout(0.25)(x)

    # BLOCK 3: Conv×2 → BN → ReLU → MaxPool → Dropout
    x = layers.Conv2D(128, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Conv2D(128, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)          # 28×28
    x = layers.Dropout(0.30)(x)

    # BLOCK 4: Conv → BN → ReLU → MaxPool → Dropout
    x = layers.Conv2D(256, (3,3), padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.MaxPooling2D((2,2))(x)          # 14×14
    x = layers.Dropout(0.30)(x)

    # FC HEAD
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(256)(x)
    x = layers.BatchNormalization()(x)
    x = layers.Activation('relu')(x)
    x = layers.Dropout(0.50)(x)

    # Output: Softmax cho 3 lớp
    outputs = layers.Dense(num_classes, activation='softmax', name='output')(x)

    return keras.Model(inputs, outputs, name='PotatoBlightCNN')


model = build_potato_cnn()
model.summary()
print(f"\n📐 Tổng tham số: {model.count_params():,}")

# ============================================================
# CÂU 3: HUẤN LUYỆN VÀ ĐÁNH GIÁ
# ============================================================
# CELL 6: Compile + Train
print("\n" + "="*60)
print("CÂU 3: HUẤN LUYỆN")
print("="*60)

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
print("✅ Loss     : Categorical Cross-Entropy")
print("✅ Optimizer: Adam (lr=0.001)")

callbacks = [
    EarlyStopping(monitor='val_loss', patience=10,
                  restore_best_weights=True, verbose=1),
    ReduceLROnPlateau(monitor='val_loss', factor=0.5,
                      patience=5, min_lr=1e-6, verbose=1),
    ModelCheckpoint('/content/best_potato_model.keras',
                    monitor='val_accuracy', save_best_only=True, verbose=1)
]

print("\n🚀 Bắt đầu huấn luyện...")
history = model.fit(
    train_generator,
    epochs=EPOCHS,
    validation_data=val_generator,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

model.save('/content/potato_blight_model.keras')
print("✅ Đã lưu model!")

# CELL 7: Vẽ đường cong huấn luyện
ep = range(1, len(history.history['loss'])+1)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].plot(ep, history.history['loss'],     'b-o', lw=2, ms=4, label='Train Loss')
axes[0].plot(ep, history.history['val_loss'], 'r-s', lw=2, ms=4, label='Val Loss')
axes[0].set_title('Loss Curve', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Loss')
axes[0].legend(); axes[0].grid(alpha=0.3)

axes[1].plot(ep, history.history['accuracy'],     'b-o', lw=2, ms=4, label='Train Accuracy')
axes[1].plot(ep, history.history['val_accuracy'], 'r-s', lw=2, ms=4, label='Val Accuracy')
axes[1].set_title('Accuracy Curve', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Accuracy')
axes[1].legend(); axes[1].grid(alpha=0.3)

plt.suptitle('Đường Cong Huấn Luyện – Potato Blight CNN',
             fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('/content/training_curves.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Đã lưu: training_curves.png")

# CELL 8: Đánh giá trên tập Test
print("\n📊 Đánh giá trên tập Test...")
test_generator.reset()
y_pred_probs = model.predict(test_generator, verbose=1)
y_pred = np.argmax(y_pred_probs, axis=1)
y_true = test_generator.classes
cls_names_eval = list(test_generator.class_indices.keys())

acc       = accuracy_score(y_true, y_pred)
precision = precision_score(y_true, y_pred, average='weighted')
recall    = recall_score(y_true, y_pred, average='weighted')

print(f"\n{'='*45}")
print(f"  KẾT QUẢ ĐÁNH GIÁ – TẬP TEST")
print(f"{'='*45}")
print(f"  Accuracy  : {acc*100:.2f}%")
print(f"  Precision : {precision*100:.2f}%")
print(f"  Recall    : {recall*100:.2f}%")
print(f"{'='*45}")
print("\n📋 Classification Report:")
print(classification_report(y_true, y_pred, target_names=cls_names_eval))

# Confusion Matrix
cm = confusion_matrix(y_true, y_pred)
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=cls_names_eval, yticklabels=cls_names_eval,
            linewidths=0.5, annot_kws={'size':13,'weight':'bold'})
ax.set_title('Confusion Matrix – Tập Test', fontsize=14, fontweight='bold', pad=15)
ax.set_ylabel('Nhãn Thực Tế', fontsize=12)
ax.set_xlabel('Nhãn Dự Đoán', fontsize=12)
plt.xticks(rotation=15); plt.yticks(rotation=0)
plt.tight_layout()
plt.savefig('/content/confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Đã lưu: confusion_matrix.png")

# Tải kết quả về máy
from google.colab import files
for f in ['/content/potato_blight_model.keras',
          '/content/training_curves.png',
          '/content/confusion_matrix.png',
          '/content/class_distribution.png']:
    files.download(f)
print("\n🎉 Hoàn thành toàn bộ!")
```
